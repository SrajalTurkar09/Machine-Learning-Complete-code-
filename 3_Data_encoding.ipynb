```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
df = pd.read_csv('/content/fictional_character_battles_complex.csv')
df
```

```python
# lets starts encoding the data
```

```python
# what is encoding
# encoding is a process of converting a variable into a number
```

```python
# Type of encoding
# 1. Label encoding
# 2. One-hot encoding
# 3. Ordinal encoding
# 4. Binary encoding
# 5. count encoding
# 6. Target encoding
```

```python
# one-hot encoding
# How it works

# we have to perform one-hot encoding for universes

from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()

df_universe = pd.DataFrame(onehotencoder.fit_transform(df[['Universe']]).toarray())
df_universe
```

```python
df.head()
```

```python
# Dummy encoding
# How it works
# it removes last 1 columns

# we have to perform dummy encoding for SpecialAbilities

# use 0 and 1 not true or false

special_abilities = pd.get_dummies(df['SpecialAbilities'])
special_abilities
```

```python
df.head()
```

```python
# label encoding

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

df_weakness = pd.DataFrame(labelencoder.fit_transform(df['Weaknesses']))
df_weakness
```

```python
df.head()
```

```python
# ordinal encoding

from sklearn.preprocessing import OrdinalEncoder
ordinalencoder = OrdinalEncoder()

df_universe = pd.DataFrame(ordinalencoder.fit_transform(df[['Universe']]))
df_universe
```

```python
# Binary encoding

from sklearn.preprocessing import Binarizer
binarizer = Binarizer(threshold = 5)

df_strength = pd.DataFrame(binarizer.fit_transform(df[['Strength']]))
df_strength
```

```python
# count encoding

from sklearn.preprocessing import CountEncoder
countencoder = CountEncoder()

df_universe = pd.DataFrame(countencoder.fit_transform(df[['Universe']]))
df_universe
```
