# 1. Agglomerative Hierarchical Clustering
    Using Scikit-Learn


```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=0.60, random_state=0)

# Apply Agglomerative Clustering
model = AgglomerativeClustering(n_clusters=3)
labels = model.fit_predict(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.title('Agglomerative Hierarchical Clustering')
plt.show()
```

# 2. Dendrogram Plotting
    Using Scipy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate sample data
np.random.seed(0)
X = np.random.rand(10, 2)

# Compute the linkage matrix
Z = linkage(X, method='ward') # ward

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

# 3. Distance Metrics and Linkage Criteria
    Using Scipy for Different Linkage Criteria

```python
from scipy.cluster.hierarchy import linkage, fcluster
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
X = np.random.rand(10, 2)

# Different linkage methods
linkage_methods = ['single', 'complete', 'average', 'ward']

for method in linkage_methods:
    Z = linkage(X, method=method)

    # Create clusters
    clusters = fcluster(Z, t=2, criterion='maxclust')

    # Plot results
    plt.figure(figsize=(8, 5))
    plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
    plt.title(f'Linkage Method: {method}')
    plt.show()
```

# 4. Hierarchical Clustering with Scikit-Learn
    Visualizing Clusters

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate sample data
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=0.60, random_state=0)

# Apply Agglomerative Clustering
model = AgglomerativeClustering(n_clusters=3)
labels = model.fit_predict(X)

# Plot results
plt.figure(figsize=(12, 6))

# Plot clusters
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.title('Agglomerative Clustering')

# Plot dendrogram
plt.subplot(1, 2, 2)
Z = linkage(X, method='ward')
dendrogram(Z)
plt.title('Dendrogram')

plt.show()
```

# 5. Hierarchical Clustering with Custom Distance Metric


```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
X = np.random.rand(10, 2)

# Compute pairwise distances
distances = pdist(X, metric='euclidean')

# Convert distance matrix to square form
distance_matrix = squareform(distances)

# Perform hierarchical clustering with custom distance matrix
Z = linkage(distance_matrix, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram with Custom Distance Metric')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```
