```python
class GradientDescent:
    def __init__(self, learning_rate=0.1, max_iters=1000, tolerance=1e-6):
        self.learning_rate = learning_rate
        self.max_iters = max_iters
        self.tolerance = tolerance

    def minimize(self, func, gradient_func, initial_guess):
        x = initial_guess

        for i in range(self.max_iters):
            gradient = gradient_func(x)
            x_new = x - self.learning_rate * gradient

            # Stopping criteria: if gradient is close to zero
            if abs(gradient) < self.tolerance:
                break

            x = x_new

            # Print current iteration information
            print(f"Iteration {i+1}: x = {x:.6f}, f(x) = {func(x):.6f}")

        print(f"\nOptimal solution: x = {x:.6f}, f(x) = {func(x):.6f}")
        return x

# Define the function f(x) = x^2 and its gradient function
def f(x):
    return x**2

def gradient_f(x):
    return 2 * x

# Parameters
learning_rate = 0.1
max_iters = 100
initial_guess = 5.0

# Create GradientDescent object
gd = GradientDescent(learning_rate, max_iters)

# Run gradient descent
optimal_x = gd.minimize(f, gradient_f, initial_guess)
print(f"\nOptimal solution: x = {optimal_x:.6f}")
```
