# Simple Linear Regression

**formula** :- y = mx + b



where,

     1. y = dependent variable

     2. x = independent variable

     3. m = slope of the line indicating the rate of change of y with respect to x

     4. b = b is value of intercept , which is the value of y when x = 0


**What is M and b **

* Slope(m) = *The slope m tells us how much the dependent variable y chsngrd for one unit increase in the independent variable x. It measures the stepness of the line.*

* Intercept(b) = *The intercept b is the value of y when = 0. It represents the starting point of the line on the y-axis*


* Best fit line = *The best fit line is that line that minimize the sum of squared difference b/w the obsereved y values annd the values predicted by the line (y = mx +b ). It represents the line that best fits the data points, showing the overall trend in the data*



```python
# Load dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Salary Data.csv")
df.head()
```

```python
# Task we have to predict the salary based on years of experience
```

```python

class LinearReg:
    def __init__(self, x, y):
        self.x = np.array(x)
        self.y = np.array(y)
        self.n = len(x)
        self.m = 0
        self.b = 0

    def cal_mean_x_y(self):
        self.x_mean = np.mean(self.x)
        self.y_mean = np.mean(self.y)
        return self.x_mean, self.y_mean

    def cal_m_b(self):
        numerator = np.sum((self.x - self.x_mean) * (self.y - self.y_mean))
        denominator = np.sum((self.x - self.x_mean) ** 2)
        self.m = numerator / denominator
        self.b = self.y_mean - self.m * self.x_mean
        return self.m, self.b

    def cal_lr(self):
        y_pred = self.m * self.x + self.b
        error = y_pred - self.y
        ss_error = np.sum(error ** 2)
        ss_total = np.sum((self.y - self.y_mean) ** 2)
        r2 = 1 - (ss_error / ss_total)
        return r2

    def predict(self, x):
        return self.m * x + self.b

    def fit(self):
        self.cal_mean_x_y()
        self.cal_m_b()

    def display(self):
        print("m =", self.m)
        print("b =", self.b)
```

```python
# apply the LR class on our data

cl = LinearReg(x=df['YearsExperience'],y=df['Salary'])
cl.fit()
cl.display()
```

```python
# prediction on new values
cl.predict(x=2.0)
```

```python
# Sklearn lenear regression

from sklearn.linear_model import LinearRegression


lr = LinearRegression()
lr.fit(df[['YearsExperience']],df['Salary'])
print(lr.coef_)
print(lr.intercept_)

# make prediction
print(lr.predict([[2.0]]))
```
