# Polynomial Regression

```python
# Polynomial Regression

# load dataset

import pandas as pd
import numpy as np
df = pd.read_csv('/content/Student_Performance.csv')
df.head()
```

```python
from os import X_OK
def polynomial_regression(X , y , degree):
  # 1. matrix formulation
  # transporse metrix
  # xTx matrix
  trans_x = np.transpose(X)
  # xTy
  trans_x_y = np.transpose(y)

  # inverse matrix of xTx
  xTx = np.linalg.inv(np.dot(trans_x , X))

  # calculate coefficients
  beta = np.dot(np.dot(xTx , trans_x) , trans_x_y)

  # predict
  y_pred = np.dot(X , beta)

  return xTx , beta , y_pred



```

```python
# encode categorical data

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Extracurricular Activities'] = le.fit_transform(df['Extracurricular Activities'])
df.head()
```

```python
# apply the model
degree = 2
X = df.drop(['Performance Index'] , axis = 1)
y = df['Performance Index']
polynomial_regression(X , y , degree)
```

```python
# sklearn polynomial regression'

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
lin_reg = LinearRegression()
lin_reg.fit(X_poly , y)
lin_reg.coef_
lin_reg.intercept_
y_pred = lin_reg.predict(X_poly)
y_pred
```
