* Normalization is a data preprocessing step used to transform the distorting differences in the ranges of values without losing information

 2. Z-score normalization
 3. RObust scaling

```python
# applying normalization

# min-max sccaler from scratch

def min_max_scaling(df, col_name):
    min_val = df[col_name].min()
    max_val = df[col_name].max()
    df[col_name] = (df[col_name] - min_val) / (max_val - min_val)
    return df

normalized_df = min_max_scaling(df, 'Age')
normalized_df

# we use age column because the distributiion of age is gaussian and min max scaler was work well on gaussian distribution
```

```python
# min-max sccaler from sklearn
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df['Age'].values.reshape(-1, 1))
df
```

```python
# Z-score normalization
# where used
# Z-score is used to normalize the data where the data is not normally distributed
# there age colomn is  normally distributed

def z_score_normalization(df, col_name):
    mean = df[col_name].mean()
    std = df[col_name].std()
    df[col_name] = (df[col_name] - mean) / std
    return df

normalized_df = z_score_normalization(df, 'Age')
normalized_df
```

```python
# z-score normalization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Age'] = scaler.fit_transform(df['Age'].values.reshape(-1, 1))
df
```

```python
# Robust scaling

# make box plot on all columns
# df.boxplot()

# there fare col has outliers

def robust_scaling(df, col_name):
  median = df[col_name].median()
  q1 = df[col_name].quantile(0.25)
  q3 = df[col_name].quantile(0.75)
  IQR = q3 - q1
  # IQR[IQR == 0] = 1e-9

  df[col_name] = (df[col_name] - median) / IQR
  return df



df = robust_scaling(df, 'Fare')
df
```

```python
# robust scaling
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df['Fare'] = scaler.fit_transform(df['Fare'].values.reshape(-1, 1))
df
```

```python
# Decimal scaling
from sklearn.preprocessing import Normalizer
scaler = Normalizer()
df['Fare'] = scaler.fit_transform(df['Fare'].values.reshape(-1, 1))
df
```

```python
# log scaling on sibsp
# beecause the sibsp has skewed distribution

def log_scaling(df, col_name):
    df[col_name] = np.log(df[col_name])
    return df

df = log_scaling(df, 'Parch')
df
```

```python
# L2 norm

# this normalization not works if data has missing values
from sklearn.preprocessing import Normalizer
scaler = Normalizer(norm='l2')
df['Age'] = scaler.fit_transform(df['Age'].values.reshape(-1, 1))
df
```
